```python
import csv
csv_list = []
with open('2023-1-8-18.txtformal.csv', 'r') as csv_file:
    reader = csv.reader(csv_file)
    csv_list = list(reader)
    del csv_list[0]


list2 = np.concatenate(csv_list).tolist()
formal_hashlist = [item for item in list2 if item != ''  ]

from collections import Counter


#1~100위 스타일 분포
formal_cate=np.array(list(set(formal_hashlist)))
array_style = np.array(formal_hashlist)
formalhash_dict=Counter(array_style)
list_key=[]
list_value = []
for key,value in formalhash_dict.items():
    list_key.append(key)
    list_value.append(value)

n = time.localtime()
s = f'{n.tm_year}-{n.tm_mon}-{n.tm_mday}-{n.tm_sec}'
f_dir = "c:\\temp\\" +s+".txt"
fc_name = f_dir+s+' formalhash.csv'

df_formalhash= pd.DataFrame(list_value,index=list_key,columns=['개수'])
ref_formal_df=df_formalhash['개수'].sort_values(ascending =False)
ref_formal_df.to_csv(fc_name,encoding='cp949') 
```

```python
df=pd.read_csv('2023-1-8-49.txt2023-1-8-49 formalhash.csv',encoding='cp949',index_col=0)
formal_fit_df=df.loc[['테이퍼드핏','루즈핏','와이드핏','오버핏','드로우핏','벌룬핏','슬림핏']]
formal_products_df = df.loc[['반소매 티셔츠','캡야구 모자','데님 팬츠','캔버스단화','셔츠블라우스']]
formal_fit_df.sort_values('개수',ascending=False,inplace=True)
formal_products_df.sort_values('개수',ascending=True,inplace=True)
```
